# ★ Hugging face 없이 csv 바로 튜닝데이터로 변환 ★ HF 호환

```python
# Define paths and filenames
fileName = "custom_hospital_situation1"
dataset_Path = "/content/gdrive/MyDrive/밸류업 프로젝트(24 여름)/hk code 다시 따라하기"
csv_file_path = os.path.join(dataset_Path, f"{fileName}.csv")
```

```python
# Load Data
try:
    df = pd.read_csv(csv_file_path, encoding="ms949")
except Exception as e:
    df = pd.read_csv(csv_file_path, encoding="utf-8")
```

```python
# 빈컬럼 생성 및 컬럼 재정의
df["input"]=""
df.columns = ["instruction","output","input"]
df.head(2)
```

```python
dataset = Dataset.from_pandas(df)
```
