```python
from huggingface_hub import HfApi
api = HfApi()

# gguf 파일 경로
file_path = "/content/gdrive/MyDrive/밸류업 프로젝트(24 여름)/eeve모델 파인튜닝/final_eeve_model_from_HF.gguf"

# Hugging Face 저장소 이름
repo_name = "sangthree/final_eeve_gguf"

# 파일 업로드
api.upload_file(
    path_or_fileobj=file_path,
    path_in_repo="final_eeve_model_from_HF.gguf",  # 저장소 내 파일 경로
    repo_id=repo_name,
    repo_type="model",  # 모델 저장소에 업로드
)
```
